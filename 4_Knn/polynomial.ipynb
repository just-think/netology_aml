# Полиномиальная регрессия

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = list(range(5))
y = np.sin(x)

plt.plot(x, y, 'o')
```

```python
x = list(np.linspace(0, 5, 7))
y = np.sin(x)

plt.plot(x, y, 'o')
```

```python
coefs = np.polyfit(x, y, 2)
coefs
```

```python
np.polyval(coefs, x)
```

```python
plt.plot(x, y, 'o')
plt.plot(x, np.polyval(np.polyfit(x, y, 2), x))
```

### Упражнение
Отобразите на одном графике результаты для степеней от 0 до 4. Посчитайте функцию потерь для каждого случая

```python
plt.plot(x, y, 'o')
for i in range(0,5):
    coefs = np.polyfit(x, y, i)
    predict = np.polyval(coefs, x)
    loss = sum((y-predict)**2)
    print(loss)
    plt.plot(x, np.polyval(np.polyfit(x, y, i), x))
```
